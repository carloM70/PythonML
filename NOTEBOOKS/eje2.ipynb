```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,5,1000)
y = x**2-2**x
plt.title('GRAFICA DE x vs. y')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y,'-r')
#plt.axis([0,2,-0.5,0.5])
plt.grid()
```

```python
theta = np.random.randn()
J = 0.5*(theta**2-2**theta)**2
alpha = 0.1
niter=100
theta_0 = []
J_0 = []
i=np.arange(niter)
for iter in range(niter):
    theta = theta - alpha*((theta**2-2**theta)*(2*theta-(2**theta)*np.log(2)))
    J = 0.5*(theta**2-2**theta)
    theta_0.append(theta)
    J_0.append(J)
plt.title('theta vs. J')
plt.xlabel('theta')
plt.ylabel('Funcion de costo J(theta)')
plt.plot(theta_0,J_0,'-r')
plt.grid()
plt.figure()
plt.title('#iteracion vs. J')
plt.xlabel('#iter')
plt.ylabel('Funcion de costo J(theta)')
plt.plot(i,J_0,'-r')
plt.grid()
print (theta)
```
