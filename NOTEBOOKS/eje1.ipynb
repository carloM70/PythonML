```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,5,1000)
y = 2*x**2-3*x+1
plt.plot(x,y,'-r')
plt.axis([0,2,-0.5,0.5])
plt.grid()
```

```python
theta = np.random.rand()
J = 0.5*(2*theta**2-3*theta+1)**2
alpha = 0.1
niter=100
theta_0 = []
J_0 = []
i=np.arange(niter)
for iter in range(niter):
    theta = theta - alpha*((2*theta**2-3*theta+1)*(4*theta-3))
    J = 0.5*(2*theta**2-3*theta+1)**2
    theta_0.append(theta)
    J_0.append(J)
plt.title('GRAFICA DE PARAMETROS VS. FUNCION DE COSTO')
plt.xlabel('PARAMETRO Theta = X')
plt.ylabel('FUNCION DE COSTO J')
plt.plot(theta_0,J_0,'-r')
plt.figure()
plt.title('GRAFICA DE NRO. ITERACIONES VS. FUNCION DE COSTO')
plt.xlabel('NRO. ITERACIONES')
plt.ylabel('FUNCION DE COSTO J')
plt.plot(i,J_0,'-r')
print ('SOLUCION DE LA ECUACION:  ',theta)
```
